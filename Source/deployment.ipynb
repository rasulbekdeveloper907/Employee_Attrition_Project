# 4️⃣ API Fundamentals

## 4.1 API nima?

API (Application Programming Interface) — ikki dastur yoki xizmatning bir-biri bilan ma'lumot almashish usuli.

Masalan:

- Frontend → Backend
- Mobile App → Server
- Backend → ML Model

---

## 4.2 REST API

REST (Representational State Transfer) — HTTP protokoli asosida ishlaydigan eng mashhur API arxitekturasi.

Afzalliklari:
- Oddiy
- Tez
- Platformadan mustaqil
- JSON bilan ishlaydi

---

## 4.3 HTTP Methods

### GET
Ma'lumot olish uchun ishlatiladi.

### POST
Yangi ma'lumot yuborish yoki prediction qilish uchun ishlatiladi.

### PUT
Mavjud ma'lumotni to'liq yangilaydi.

### PATCH
Ma'lumotning faqat kerakli qismini yangilaydi.

### DELETE
Ma'lumotni o'chiradi.

---

## 4.4 Request va Response

**Request** — mijoz (client) serverga yuboradigan so'rov.

**Response** — server qaytaradigan javob.

---

## 4.5 HTTP Status Codes

- **200 OK** — So'rov muvaffaqiyatli bajarildi.
- **201 Created** — Yangi resurs yaratildi.
- **400 Bad Request** — Noto'g'ri so'rov yuborildi.
- **401 Unauthorized** — Autentifikatsiya talab qilinadi.
- **403 Forbidden** — Ruxsat yo'q.
- **404 Not Found** — Resurs topilmadi.
- **500 Internal Server Error** — Server ichki xatosi.

---

## 4.6 JSON

JSON (JavaScript Object Notation) — API orqali ma'lumot almashishning eng ommabop formati.

Xususiyatlari:
- Yengil format
- O'qish oson
- Tilga bog'liq emas
- FastAPI va boshqa backend frameworklar bilan to'liq mos
