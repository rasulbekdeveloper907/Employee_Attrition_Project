# Docker nima?

Docker — bu ilovalarni (Application) ularning barcha kerakli kutubxonalari, fayllari va sozlamalari bilan birga bitta paketga joylab, istalgan kompyuterda bir xil ishlashini ta'minlaydigan platforma.

Oddiy qilib aytganda:

> **"Docker — bu dasturingizni qadoqlab, hamma joyda bir xil ishlashini ta'minlaydigan vosita."**

---

# Docker nima uchun kerak?

ML loyihasida sizning kompyuteringizda quyidagilar bo'lishi mumkin:

- Python 3.13
- FastAPI
- Pandas
- NumPy
- Scikit-learn
- XGBoost

Lekin boshqa kompyuterda:

- Python 3.10
- Pandas eski versiya
- XGBoost umuman o'rnatilmagan

Natijada loyiha ishlamasligi mumkin.

Docker bu muammoni hal qiladi.

---

# Docker bo'lmasa

```text
Developer PC
│
├── Python 3.13
├── FastAPI
├── Pandas
├── XGBoost
└── Project

        ↓

Server

├── Python yo'q
├── Pandas eski
├── XGBoost yo'q

❌ Loyiha ishlamaydi
```

---

# Docker bilan

```text
Docker Image
│
├── Python
├── FastAPI
├── Pandas
├── XGBoost
├── Model.pkl
└── Project

        ↓

Istalgan kompyuter

        ↓

✅ Hammasi bir xil ishlaydi
```

---

# Docker qaysi muammoni hal qiladi?

Docker quyidagi muammolarni bartaraf etadi:

- "Menda ishlaydi, serverda ishlamaydi."
- Kutubxona versiyalari mos emas.
- Python versiyasi boshqacha.
- Sozlamalar turlicha.
- Deployment murakkab.

# Dockerfile nima?

Dockerfile — bu Docker Image yaratish uchun yoziladigan oddiy matnli fayl.

Unda Docker'ga quyidagi buyruqlar beriladi:

- Qaysi Python versiyasidan foydalanish
- Ishchi papkani yaratish
- Fayllarni nusxalash
- Kutubxonalarni o'rnatish
- Serverni ishga tushirish

Oddiy qilib aytganda,

Dockerfile — bu **Docker uchun yo'riqnoma**.

---

# Dockerfile qayerda bo'ladi?

Loyiha tuzilishi:

```text
Employee_Attrition_Project/
│
├── main.py
├── requirements.txt
├── Dockerfile
├── Models/
│   └── xgboost_model.pkl
└── README.md
```

**Eslatma:** Fayl nomi aynan `Dockerfile` bo'ladi. Kengaytmasi (`.txt`) bo'lmaydi.

---

# Dockerfile qanday ishlaydi?

```text
Dockerfile
      │
      ▼
docker build
      │
      ▼
Docker Image
      │
      ▼
docker run
      │
      ▼
Container
```

---

# Dockerfile sintaksisi

Dockerfile qatorma-qator bajariladi.

Misol:

```dockerfile
FROM python:3.13

WORKDIR /app

COPY . .

RUN pip install -r requirements.txt

CMD ["python", "main.py"]
```

Docker har bir qatorni ketma-ket bajaradi.

---

# 1. FROM

FROM — qaysi asosiy Image ishlatilishini belgilaydi.

Masalan:

```dockerfile
FROM python:3.13
```

Bu:

```
Docker Hub

↓

python:3.13

↓

Image
```

---

Yana misollar:

```dockerfile
FROM ubuntu:24.04
```

```dockerfile
FROM node:22
```

```dockerfile
FROM nginx
```

---

# 2. WORKDIR

Container ichidagi ishchi papkani belgilaydi.

```dockerfile
WORKDIR /app
```

Natija:

```
Container

↓

/app
```

Endi barcha buyruqlar `/app` ichida bajariladi.

---

# 3. COPY

Fayllarni Container ichiga nusxalaydi.

```dockerfile
COPY . .
```

Birinchi nuqta:

```
Lokal loyiha
```

Ikkinchi nuqta:

```
Container (/app)
```

Natija:

```
Employee_Project/

↓

Container

/app
```

---

Misol:

```dockerfile
COPY requirements.txt .
```

Faqat bitta faylni nusxalaydi.

---

# 4. RUN

Image yaratilayotgan vaqtda buyruq bajaradi.

Masalan:

```dockerfile
RUN pip install -r requirements.txt
```

Bu barcha kutubxonalarni o'rnatadi.

---

Yana misollar:

```dockerfile
RUN mkdir data
```

```dockerfile
RUN apt-get update
```

---

# RUN va CMD farqi

RUN

↓

Image build vaqtida ishlaydi.

CMD

↓

Container ishga tushganda ishlaydi.

---

# 5. EXPOSE

Container qaysi portni ishlatishini bildiradi.

FastAPI odatda:

```dockerfile
EXPOSE 8000
```

Bu:

```
Container

↓

8000-port
```

---

# 6. CMD

Container ishga tushganda bajariladigan asosiy buyruq.

FastAPI uchun:

```dockerfile
CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
```

Bu:

```
uvicorn

↓

main.py

↓

app

↓

8000-port
```

---

# To'liq Dockerfile

Sizning Employee Attrition API loyihangiz uchun:

```dockerfile
FROM python:3.13

WORKDIR /app

COPY . .

RUN pip install --no-cache-dir -r requirements.txt

EXPOSE 8000

CMD ["uvicorn", "main:app", "--host", "0.0.0.0", "--port", "8000"]
```

Docker Container vaqtinchalik (Temporary).
