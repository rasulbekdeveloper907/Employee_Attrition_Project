# FastAPI nima?

**FastAPI** — bu **Python** dasturlash tilida yozilgan zamonaviy, tezkor va yuqori unumdor (high-performance) **Web Framework** bo'lib, REST API va backend servislarni yaratish uchun ishlatiladi.

FastAPI 2018-yilda **Sebastián Ramírez** tomonidan yaratilgan.

U **ASGI (Asynchronous Server Gateway Interface)** standartiga asoslangan bo'lib, `Starlette` va `Pydantic` kutubxonalaridan foydalanadi.

---

# FastAPI ning rasmiy ta'rifi

> FastAPI — bu Python 3.8+ uchun type hints asosida API yaratishga mo'ljallangan zamonaviy, tezkor web framework.

---

# FastAPI nima uchun yaratilgan?

Oldingi frameworklarda ko'p muammolar bor edi.

Masalan:

- Juda ko'p kod yozish kerak edi.
- Validation qo'lda bajarilardi.
- Documentation qo'lda yozilardi.
- Async dasturlash murakkab edi.
- Performance pastroq edi.

FastAPI ushbu muammolarni hal qilish uchun yaratilgan.

---

# FastAPI qanday Framework?

FastAPI — **Backend Web Framework**.

U orqali quyidagilarni yaratish mumkin:

- REST API
- AI API
- ML Model Deployment
- Authentication System
- CRUD API
- Microservices
- Real-time API
- WebSocket Server
- File Upload API
- Cloud API
